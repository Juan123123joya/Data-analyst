```python
import pandas as pd

top_countries = df['country'].value_counts().nlargest(10).index

filtered_df = df[df['country'].isin(top_countries)]

print(filtered_df)

```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y="country", data=filtered_df, palette="coolwarm", order=filtered_df['country'].value_counts().index)
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.title("Top 10 Countries with the Most Netflix Titles")
plt.show()
```

```python
country_counts = filtered_df['country'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("coolwarm", len(country_counts)))
plt.title("Distribution of Netflix Titles by Country (Top 10)")
plt.axis('equal')  # Igualar el aspecto
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_df, x='release_year', y='country', hue='country', palette="coolwarm", alpha=0.7)
plt.xlabel("Release Year")
plt.ylabel("Country")
plt.title("Netflix Titles Release Year by Country (Top 10)")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
year_counts = filtered_df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Number of Netflix Titles Released Over the Years (Top 10 Countries)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='rating', data=df, palette='coolwarm', order=df['rating'].value_counts().index)
plt.xlabel('Number of Titles')
plt.ylabel('Rating')
plt.title('Number of Titles by Rating')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=df, palette='coolwarm')
plt.title('Conteo de Títulos por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Número de Títulos')
plt.show()
```

```python
top_directors = df['director'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='plasma')
plt.title('Top 10 Directores con Más Títulos')
plt.xlabel('Número de Títulos')
plt.ylabel('Director')
plt.show()
```
