```python
import requests
import pandas as pd

# Clave de API de Financial Modeling Prep
api_key = 'xZlhWmlvSXJyg1eSXyPlll2Q1Q0vzgPZ'

# Obtener la lista de compañías
url = f"https://financialmodelingprep.com/api/v3/stock/list?apikey={api_key}"
response = requests.get(url)

if response.status_code == 200:
    stock_list = response.json()
    symbols = [stock["symbol"] for stock in stock_list[:10]]  # Tomamos las primeras 10 acciones para prueba
else:
    print("Error al obtener la lista de acciones")
```

```python
# Diccionario para almacenar los datos
stock_data = {}

# Obtener datos para cada acción
for symbol in symbols:
    url = f"https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?apikey={api_key}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if "historical" in data:
            df = pd.DataFrame(data["historical"])
            df["date"] = pd.to_datetime(df["date"])
            df.set_index("date", inplace=True)
            df = df.sort_index()  # Ordenar por fecha
            stock_data[symbol] = df
    else:
        print(f"Error al obtener datos de {symbol}")
```

```python
for symbol, df in stock_data.items():
    df["Returns"] = df["close"].pct_change()
    df["Cumulative Returns"] = (1 + df["Returns"]).cumprod()

# Mostrar rentabilidad acumulada de una acción de ejemplo
print(stock_data[symbols[0]][["close", "Returns", "Cumulative Returns"]].tail())
```

```python
ranked = sorted(stock_data.items(), key=lambda x: x[1]["Cumulative Returns"].iloc[-1], reverse=True)

print("Acciones más rentables:")
for symbol, df in ranked:
    print(f"{symbol}: Rentabilidad total {df['Cumulative Returns'].iloc[-1] - 1:.2%}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
for symbol, df in ranked[:5]:  # Graficamos las 5 mejores acciones
    plt.plot(df.index, df["Cumulative Returns"], label=symbol)

plt.title("Rentabilidad Acumulada de las Acciones Más Rentables")
plt.xlabel("Fecha")
plt.ylabel("Rentabilidad Acumulada")
plt.legend()
plt.show()
```
